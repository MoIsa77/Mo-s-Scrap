```python
from selenium import webdriver
import pandas as pd
import time
from tabulate import tabulate
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
link = "https://www.formula1.com/en/racing/2023.html"
driver = webdriver.Chrome()
driver.get(link)

# RUN LEBIH DARI 1X UNTUK HASIL YANG MAKSIMAL !!!

# Ini Tombol Auto Klik
accept_button_locator = (By.XPATH, "//button[@id='truste-consent-button']")
wait = WebDriverWait(driver, 10)
accept_button = wait.until(EC.element_to_be_clickable(accept_button_locator))
accept_button.click()

# Schedule Scrap
Action_Chains = ActionChains(driver)
items = driver.find_elements(By.XPATH, "//*[contains(@class, 'col-12 col-sm-6 col-lg-4 col-xl-3')]")
Class_Catch_1 = []
for item in items:
    round = item.find_element(By.XPATH, ".//*[contains(@class, 'card-title') and contains(@class, 'f1-uppercase') and contains(@class, 'f1-color--warmRed')]").text
    country = item.find_element(By.XPATH, ".//*[contains(@class, 'event-place')]").text
    times_date = item.find_element(By.XPATH, ".//*[contains(@class, 'no-margin')]").text
    times_month = item.find_element(By.XPATH, ".//*[contains(@class, 'month-wrapper f1-wide--xxs')]").text.replace('\ ',' ')
    circuit_name = item.find_element(By.XPATH, ".//*[contains(@class, 'event-title f1--xxs')]").text

    Scroll_1 = item.find_element(By.XPATH, ".//*[contains(@class, 'event-item-wrapper event-item-link')]")
    Action_Chains.move_to_element(Scroll_1).perform()

    Class_Catch_1.append((round, country, times_date, times_month, circuit_name))

print(Class_Catch_1)

df_1 = pd.DataFrame(Class_Catch_1, columns=['ROUND','COUNTRY','DATE','MONTH','CIRCUIT'])

driver.quit()

# Teams Scrap
link_2 = "https://www.formula1.com/en/teams.html"
driver = webdriver.Chrome()
driver.get(link_2)

accept_button_locator = (By.XPATH, "//button[@id='truste-consent-button']")
wait = WebDriverWait(driver, 0)
accept_button = wait.until(EC.element_to_be_clickable(accept_button_locator))
accept_button.click()
driver.implicitly_wait(0)
driver.implicitly_wait(0)

Action_Chains_2 = ActionChains(driver)
Class_Catch_2 = []
items = driver.find_elements(By.XPATH, "//*[contains(@class, 'col-12') and contains(@class, 'col-md-6')]")

for item in items:
    team = item.find_element(By.XPATH, ".//*[contains(@class, 'f1-color--black')]").text
    point_ril = item.find_element(By.XPATH, ".//*[contains(@class, 'f1-wide--s')]").text
    driver_name = item.find_element(By.XPATH, ".//*[contains(@class, 'listing-team')]").text.replace('\n',' ')

    Scroll_2 = item.find_element(By.XPATH, ".//*[contains(@class, 'listing-item-wrapper')]")
    Action_Chains_2.move_to_element(Scroll_2).perform()

    Class_Catch_2.append((team, point_ril, driver_name))

print(Class_Catch_2)

df_2 = pd.DataFrame(Class_Catch_2, columns=['TEAMS', 'PTS', 'DRIVER'])
print(df_1, df_2)

driver.quit()
```

```python
# Export To Excel/XLSX
df_1.to_excel('Data_1_Results.xlsx',index=False)
print("DATA 1 EXPORT COMPLETE")

df_2.to_excel('Data_2_Results.xlsx',index=False)
print("DATA 2 EXPORT COMPLETE")
```
